```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
# 读取两张图像
img1 = plt.imread('图片\LSTM_LOSS.png')
img2 = plt.imread('图片\LSTM_MAE.png')
img3 = plt.imread('图片\LSTM_Train.png')
img4 = plt.imread('图片\LSTM_Test.png')
font = FontProperties(fname='SimHei.ttf')
# 创建一个新的图像,并将四张图像作为子图
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

# 显示四张图像
ax1.imshow(img1)
ax1.axis('off')
ax1.annotate('图7.1', xy=(0.51, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)

ax2.imshow(img2)
ax2.axis('off')
ax2.annotate('图7.2', xy=(0.51, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)

ax3.imshow(img3)
ax3.axis('off')
ax3.annotate('图7.3', xy=(0.51, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)

ax4.imshow(img4)
ax4.axis('off')
ax4.annotate('图7.4', xy=(0.525, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)
# 调整子图之间的间距
plt.subplots_adjust(wspace=0.05, hspace=0.05)

# 显示合并后的图像
plt.show()
```
