```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
# 
img1 = plt.imread('\LSTM_LOSS.png')
img2 = plt.imread('\LSTM_MAE.png')
img3 = plt.imread('\LSTM_Train.png')
img4 = plt.imread('\LSTM_Test.png')
font = FontProperties(fname='SimHei.ttf')
# ,
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(20, 15))

# 
ax1.imshow(img1)
ax1.axis('off')
ax1.annotate('7.1', xy=(0.51, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)

ax2.imshow(img2)
ax2.axis('off')
ax2.annotate('7.2', xy=(0.51, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)

ax3.imshow(img3)
ax3.axis('off')
ax3.annotate('7.3', xy=(0.51, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)

ax4.imshow(img4)
ax4.axis('off')
ax4.annotate('7.4', xy=(0.525, -0.03), xycoords='axes fraction', fontsize=12, fontproperties=font)
# 
plt.subplots_adjust(wspace=0.05, hspace=0.05)

# 
plt.show()
```
