```python
import matplotlib.pyplot as plt
import re
def plot_loss_mae_mse(output):
    # 提取损失、MAE和MSE的值
    loss_values = []
    mae_values = []
    mse_values = []
    epochs = []

    lines = output.strip().split('\n')
    for i in range(0, len(lines), 2):
        epoch_line = lines[i]
        values_line = lines[i+1]

        epoch = int(epoch_line.split('/')[0].split()[-1])
        epochs.append(epoch)

        match = re.search(r'loss: (\d+\.\d+e?-?\d*) - mae: (\d+\.\d+e?-?\d*) - mse: (\d+\.\d+e?-?\d*)', values_line)
        if match:
            loss_values.append(float(match.group(1)))
            mae_values.append(float(match.group(2)))
            mse_values.append(float(match.group(3)))

    # 绘制损失曲线
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, loss_values, label='Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training Loss')
    plt.legend()
    plt.show()

    # 绘制MAE曲线  
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, mae_values, label='MAE')
    plt.xlabel('Epoch')
    plt.ylabel('MAE')
    plt.title('Training Mean Absolute Error')
    plt.legend()
    plt.show()

    # 绘制MSE曲线
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, mse_values, label='MSE') 
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.title('Training Mean Squared Error')
    plt.legend()
    plt.show()
```

```python
# 读取output.txt文件内容作为output变量
with open('output.txt', 'r') as f:
    output = f.read()
plot_loss_mae_mse(output)    
```
